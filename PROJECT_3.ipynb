## SIMPLE IMPUTER

```python
from sklearn.impute import SimpleImputer
```

```python
#FOR CONTINUOUS COLUMNS
cont_imp = SimpleImputer(strategy="median")
df_train[["Age","Annual Income","Vehicle Age","Insurance Duration","Health Score","Credit Score"]] = cont_imp.fit_transform(df_train[["Age","Annual Income","Vehicle Age","Insurance Duration","Health Score","Credit Score"]])


#For Categorical Columns
cat_imp = SimpleImputer(strategy="most_frequent")
df_train[["Marital Status","Occupation","Customer Feedback","Previous Claims"]] = cat_imp.fit_transform(df_train[["Marital Status","Occupation","Customer Feedback","Previous Claims"]])
```

## FILL WITH FORWARD FILL

```python
df_train["Number of Dependents"].fillna(method="ffill", inplace=True)
```

# AFTER CAPPING IQR METHOD

```python
df_train["income_capped"] = df_train["Annual Income"].clip(
    lower = df_train["Annual Income"].quantile(0.1),
    upper = df_train["Annual Income"].quantile(0.94))
```

```python
sns.boxplot(x=df_train["income_capped"])
```

```python
df_train["Previous Claims Capp"] = df_train["Previous Claims"].clip(
    lower = df_train["Previous Claims"].quantile(0.0),
    upper = df_train["Previous Claims"].quantile(0.948))
```

```python
sns.boxplot(x=df_train["Previous Claims Capp"])
```

## ONE HOT ENCODER

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
encoder_train = OneHotEncoder(sparse_output=False,drop=None)
data_encoded_train = encoder_train.fit_transform(df_train[["Gender","Marital Status","Occupation","Location","Smoking Status","Property Type"]])
train_encoded_columns = []
for col_name, categories in zip(["Gender","Marital Status","Occupation","Location","Smoking Status","Property Type"], encoder_train.categories_):
    train_encoded_columns.extend([f"{col_name}_{cat}" for cat in categories])
encoded_df_train = pd.DataFrame(data_encoded_train, columns = train_encoded_columns)
```

```python
encoded_df_train
```

```python
df_train
```

## ORDINAL ENCODER

```python
train_ordinal_cols = ["Education Level", "Policy Type", "Customer Feedback", "Exercise Frequency"]
train_encoded_columns = [
    ["High School", "Bachelor's", "Master's", "PhD"],            # Education Level
    ["Basic", "Comprehensive", "Premium"],                       # Policy Type
    ["Poor", "Average", "Good",],                                # Customer Feedback
    ["Rarely", "Daily", "Weekly", "Monthly"]                     # Exercise Frequency
]
```

```python
oe = OrdinalEncoder(categories=train_encoded_columns)
df_train[train_ordinal_cols] = oe.fit_transform(df_train[train_ordinal_cols].astype(str))
```

```python
df_train
```

### CHANGING DATA TYPE OF Previous Claims Capp COLUMN BECAUSE ITS IN NUMBER

```python
df_train["Previous Claims Capp"]= df_train['Previous Claims Capp'].astype(float)
```

```python
df_train.info()
```
